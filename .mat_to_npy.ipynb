```python
import scipy.io
import numpy as np
import os
import glob

MAT_DIR        = r"C:\Users\RF-LAB\Desktop\Pipeline\valid_test_iq"
OUTPUT_NPY_DIR = r"C:\Users\RF-LAB\Desktop\Pipeline\Valid_test_processed_iq"

os.makedirs(OUTPUT_NPY_DIR, exist_ok=True)
mat_files = sorted(glob.glob(os.path.join(MAT_DIR, "*.mat")))
print(f"Found {len(mat_files)} .mat file(s)")

for mat_path in mat_files:
    fname = os.path.splitext(os.path.basename(mat_path))[0]
    mat   = scipy.io.loadmat(mat_path)

    # Show keys so you can identify which variable holds IQ data
    data_keys = [k for k in mat.keys() if not k.startswith('__')]
    print(f"\n{fname} → keys: {data_keys}")
```

```python
for mat_path in mat_files:
    fname = os.path.splitext(os.path.basename(mat_path))[0]
    mat   = scipy.io.loadmat(mat_path)
    keys  = [k for k in mat.keys() if not k.startswith('__')]
    print(f"{fname} → {keys}")
```

```python
mat = scipy.io.loadmat(mat_files[0])
data = mat['previous_matrix']
print("Shape:", data.shape)
print("Dtype:", data.dtype)
print("Is complex:", np.iscomplexobj(data))
```

```python
# Conversion code cell
for mat_path in mat_files:
    fname = os.path.splitext(os.path.basename(mat_path))[0]
    mat   = scipy.io.loadmat(mat_path)

    iq = mat['previous_matrix']   # shape: (3, N) complex128

    # Use row 0 (first receiver channel) — change index if needed
    data = iq[0].astype(np.complex64)  # shape: (N,)

    out_path = os.path.join(OUTPUT_NPY_DIR, f"{fname}_processed.npy")
    np.save(out_path, data)
    print(f"  [✓] {fname}  →  shape={data.shape}  saved to {out_path}")

print(f"\nDone. {len(mat_files)} file(s) converted to .npy")
```
