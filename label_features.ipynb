```python
import pandas as pd

INPUT_CSV  = r"C:\Users\RF-LAB\Desktop\Pipeline\Features_DroneRF.csv"
OUTPUT_CSV = r"C:\Users\RF-LAB\Desktop\Pipeline\Features_DroneRF_labeled.csv"
```

```python
df = pd.read_csv(INPUT_CSV)

# DroneRF labeling rule:
# filename starts with '00000' = No-Drone (0)
# everything else              = Drone (1)
df['label'] = df['file'].str.startswith('00000').map({True: 0, False: 1})
```

```python
# Verify
print(f"Total rows       : {len(df)}")
print(f"Drone     (1)    : {(df['label']==1).sum()}")
print(f"No-Drone  (0)    : {(df['label']==0).sum()}")
print(f"\nSample no-drone files:")
print(df[df['label']==0]['file'].head(5).tolist())
print(f"\nSample drone files:")
print(df[df['label']==1]['file'].head(5).tolist())
```

```python
# Drop file column — classifier must never see filenames
df = df.drop(columns=['file'])

df.to_csv(OUTPUT_CSV, index=False)
print(f"\nSaved → {OUTPUT_CSV}")
print(f"Columns : {df.columns.tolist()}")
```
